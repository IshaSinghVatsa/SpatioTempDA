# NEW DATAFRAME WITH ONLY TWO COLUMNS

```python
new_df=df[df['WHO Region']=='Europe']
new_df_1 = new_df[['Deaths / 100 Cases','WHO Region']]
new_df_1.head()

#new data with WHO region and Deaths/100 cases
```

5 NUMBER SUMMARY

```python
new_df_1.describe()

#5 number summary of deaths/100 cases
   #min = 0
   #Q1. = 1.61
   #Q2. = 3.09
   #Q3. = 4.78
   #max = 15.19
```

```python
import seaborn as sns
sns.boxplot(data=new_df_1)
sns.swarmplot(data=new_df_1)
plt.show()

#box plot of 5 number summary
```

```python
from numpy import percentile
quartiles = percentile(new_df_1, [25,50,75])
data_min,data_max = new_df_1.min(),new_df_1.max()
print('Min: %.3f' % data_min)
print('Q1: %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3: %.3f' % quartiles[2])
print('Max: %.3f' % data_max)
```
